```python
FILE = 'process.txt'
process_dict = {}
zero = Process(0, -1)
process_dict[0] = zero
        
class Process:
    def __init__(self, pid, ppid):
        self.pid = pid
        self.ppid = ppid
        self.children = []
        
    def add_child(self, child):
        self.children.append(child)
        
    def __str__(self):
        children = [child.pid for child in self.children]
        return f'pid={self.pid}, ppid={self.ppid}, children={children}'
        
with open(FILE) as file:
    file.readline()
    for line in file:
        process_fields = line.strip().split()
        pid = int(process_fields[1])
        ppid = int(process_fields[2])
        process = Process(pid, ppid)
        process_dict[pid] = process
        process_dict[ppid].add_child(process)

for pid, process in process_dict.items():
    print(process)
```
